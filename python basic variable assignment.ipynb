```python
#Q.3 Write a python programme that converts temperature from celsius to fahrenheit. Take the temperature in celsius as input,store it in a variable,convert it into fahrenheit,and display the result.
```

```python
celsius=float(input("enter temperature in celsius:"))
```

```python
fahrenheit=(celsius *9/5) + 32
```

```python
print(f"{celsius} degrees Celsius is equal to {fahrenheit} degrees Fahrenheit.")
```
